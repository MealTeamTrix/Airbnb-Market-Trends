# Exploring Airbnb Market Trends

This is a datacamp project to answer the following questions:

* What are the dates of the earliest and most recent reviews? 
* How many of the listings are private room?
* What is the average listing price?
* Combine the created variables into a new DF called review_dates

```python
# Import libaries
import pandas as pd
import numpy as np
```

```python
# Read in the data
# Read in .csv
df_price = pd.read_csv('data/airbnb_price.csv')
# Read in .xlsx
df_type = pd.read_excel('data/airbnb_room_type.xlsx')
# Read in .tsv file
df_review = pd.read_csv('data/airbnb_last_review.tsv', sep='\t')

# View the results
df_price.head()
```

```python
df_type.head()
```

```python
df_review.head()
```

```python
# All of the Df can be joined on the `listing_id` column
merge_df = pd.merge(df_price, df_type, on='listing_id', how='left')
merge_df = pd.merge(merge_df, df_review, on='listing_id', how='left')

# View the results
merge_df.head()
```

```python
# What are the the dates of the earliest and most recent reviews?
# Check the metadata to identify the datatypes for `last_review` column
merge_df.info()

# Transform `last_review` to datetime format
merge_df['last_review'] = pd.to_datetime(merge_df['last_review'], format = '%B %d %Y')
# Identify the first review date using .min() function
first_review = merge_df['last_review'].min()
# Identify the most recent review using .max() function
last_review = merge_df['last_review'].max()

# View the results
print(f'This is the first review: {first_review}')
print(f'This is the most recent review: {last_review}')

# Find the space in time between reviews
gap = first_review - last_review
print(f'the time gap between the first and last review is: {gap} days.')
```

```python
# How many listings are private rooms?
# First look at the `room_type` column to identify its format
print(merge_df['room_type'])
# Use the str.lower() function to make all lower case
merge_df['room_type'] = merge_df['room_type'].str.lower()

# Count the number of private rooms
private_room_count = merge_df[merge_df['room_type'] == 'private room'].shape[0]

# Print the results
print(f'There are a total of: {private_room_count} private rooms.')
```

```python
# What is the average listing price?
# Look at the `price` column for potential transformation
print(merge_df['price'])

# Need to remove 'dollars' and transform from object to int
merge_df['clean_price'] = merge_df['price'].str.replace('dollars', '').astype(float)

# Find the average price using the .mean() function
avg_price = merge_df['clean_price'].mean()

# View the results
print(f'The average price of a listing is: {avg_price}.')
```

```python
# Put the data into a DF
review_dates = pd.DataFrame({
    'first_reviewed': [first_review],
    'last_reviewed': [last_review],
    'nb_private_rooms': [private_room_count],
    'avg_price': [avg_price]
})

# View the results
review_dates
```
